Business Problem:
Hotel has seen a high cancellation rates
Due to such issue, it results in loower of revenues and ideal hotel room usage
Identify other factors that have impact on cancellations

Assumptions:
There are no outliers within a span of 2 years 2015-17
The information is current and can be used to perform current analysis
The main factor that affects the revenues is due to the bookings cancellations
Cancellations results in vacant rooms 

Research Questions:
What are some of the factors that affects the hotels reservations cancellations
Suggest ways on improving hotels reservations cancellations better
Assist hotels in making price and promotional decisions

Hypothesis
More cancellations occur when prices are higher
When there is longer waiting list, customers tend top cancel more frequently
Majority of clients are coming from offline travel agents to make their reservations
